1 - What is NumPy, and why is it widely used in Python?

Ans - NumPy is a powerful library for numerical computing in Python. It provides support for arrays, matrices, and many mathematical functions. It's widely used because it's fast, supports vectorized operations, and is essential for scientific computing.

2 - How does broadcasting work in NumPy?

Ans - Broadcasting in NumPy allows operations on arrays of different shapes. Smaller arrays are 'broadcast' over larger arrays so that arithmetic operations can be performed without explicit replication.

3 - What is a Pandas DataFrame?

Ans - A Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table and is one of the most commonly used structures in data analysis.

4 - Explain the use of the groupby() method in Pandas?

Ans - The groupby() method in Pandas is used to group data based on some criteria, then apply a function like mean(), sum(), or count() to each group. It is useful for data aggregation and summarization.

5 - Why is Seaborn preferred for statistical visualizations?

Seaborn is built on top of Matplotlib and offers a high-level interface for drawing attractive and informative statistical graphics. It simplifies many tasks such as visualizing distributions and relationships.

6 - What are the differences between NumPy arrays and Python lists?

Ans - NumPy arrays are more efficient, support vectorized operations, and can be multi-dimensional. Python lists are more flexible but slower for numerical operations and lack advanced features like broadcasting.

7 - What is a heatmap, and when should it be used?

Ans - A heatmap is a graphical representation of data where individual values are represented as colors. It is used to visualize the magnitude of values across a matrix or correlation data.

8 - What does the term “vectorized operation” mean in NumPy?

Ans - Vectorized operations refer to operations that are applied to arrays element-wise without the need for explicit loops. This leads to more concise and faster code.

9 - How does Matplotlib differ from Plotly?

Ans - Matplotlib is a 2D plotting library good for static plots, while Plotly supports interactive plots, better for web apps and dashboards. Plotly also has more built-in support for complex visualizations.

10 - What is the significance of hierarchical indexing in Pandas?

Ans - Hierarchical indexing allows for multiple levels of indexing in a DataFrame, making it easier to work with higher-dimensional data in a 2D structure.

11 - What is the role of Seaborn’s pairplot() function?

Ans - Seaborn’s pairplot() is used to visualize pairwise relationships in a dataset. It plots scatter plots for each pair of features and histograms for individual features.

12 - What is the purpose of the describe() function in Pandas?

Ans - The describe() function in Pandas provides a summary of statistics (like mean, std, min, max) for numerical columns, giving an overview of the dataset.

13 - Why is handling missing data important in Pandas?

Ans - Missing data can skew analysis and lead to inaccurate conclusions. Pandas provides tools like fillna(), dropna(), and isnull() to handle missing values efficiently.

14 - What are the benefits of using Plotly for data visualization?

Ans - Plotly offers interactive, web-ready visualizations, supports dashboards, has a wide variety of chart types, and integrates well with Dash for building analytical web apps.

15 - How does NumPy handle multidimensional arrays?

Ans - NumPy provides support for n-dimensional arrays through its ndarray object, allowing easy indexing, slicing, reshaping, and mathematical operations across dimensions.

16 - What is the role of Bokeh in data visualization?

Ans - Bokeh is a Python library for creating interactive and real-time visualizations for modern web browsers. It is suitable for large streaming datasets and dashboards.

17 - Explain the difference between apply() and map() in Pandas?

Ans - apply() can be used on both Series and DataFrames and allows applying functions along an axis, while map() is only for Series and is used for element-wise transformations.

18 - What are some advanced features of NumPy?

Ans - Advanced features include broadcasting, vectorized operations, memory mapping, FFTs, linear algebra functions, masked arrays, and random sampling.

19 - How does Pandas simplify time series analysis?

Ans - Pandas provides date_range(), resample(), shift(), and time-aware indexing, making time series analysis intuitive and powerful.

20 - What is the role of a pivot table in Pandas?

Ans - Pivot tables allow data to be reshaped and summarized using aggregation functions. They are essential for reporting and multidimensional data analysis.

21 - Why is NumPy’s array slicing faster than Python’s list slicing?

Ans - NumPy arrays are stored in contiguous memory blocks and use optimized C-based implementations, making slicing operations significantly faster than with Python lists.

22 - What are some common use cases for Seaborn?

Ans - Common use cases include visualizing distributions, categorical data, regression relationships, and multivariate data using plots like boxplots, violin plots, and heatmaps.

```python
#Q2: Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [15, 25, 35]})
mean_A = df['A'].mean()
print(mean_A)
```

```python
# Q10: Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

# Create the histogram
sns.histplot(data, kde=True)  # kde=True adds a kernel density estimate
plt.title('Distribution Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
